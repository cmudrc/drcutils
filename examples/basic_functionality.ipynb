# Accessing brand colors
The `brand` module contains a lot of stuff. One of those things is the `COLORS` list, which lists the hex codes of the colors of the DRC brand. You can do things with them, like this:

```
import matplotlib.pyplot # for plotting
import drcutils.brand # get the colors from here

# Make a bar plot with the DRC colors. Labels on the x-axis are hex codes.
matplotlib.pyplot.bar(
    [1, 2, 3, 4, 5, 6], # x data for the plot
    [1, 1, 1, 1, 1, 1], # y data for the plot
    color=drcutils.brand.COLORS, # plot with the colors
    tick_label=drcutils.brand.COLORS # plot the hex name of the colors
);
```

# Using colormaps
Those colors have been worked into some other pre-built functions though, so you don't have to do everything from scratch. For instance, here's a demonstration of the `matplotlib` colormaps you can import.

```
import matplotlib.pyplot # for plotting
import numpy # for numerical stuff
from drcutils.colormaps import * # for colormaps

# Make a figure to plot on
fig = matplotlib.pyplot.figure()

# This is what we'll plot
colormaps = [
    hamster,
    hamster_r,
    warm_hamster,
    warm_hamster_r,
    cool_hamster,
    cool_hamster_r,  
    diverging_hamster,
    diverging_hamster_r,
    dark_diverging_hamster,
    dark_diverging_hamster_r,
]

# Iterate through subplots
subplot_index = 1
for cm in colormaps:
    sp = matplotlib.pyplot.subplot(5, 2, subplot_index)
    ax = sp.imshow([numpy.linspace(0, 1, 30)], cmap=cm)
    sp.set_title(cm.name)
    matplotlib.pyplot.axis('off');
    subplot_index += 1

# Make sure it looks ok
fig.set_figwidth(10)
fig.set_dpi(200)
```

# Using our logos and wordmarks
Of course, our brand is more than just colors. You can alos directly import brand images and files!

```
# Imports from DRC library
import drcutils.brand

# For plotting
import matplotlib.pyplot
import PIL

# Make a figure to plot on
fig = matplotlib.pyplot.figure()

# This is what we'll plot
brand_images = [
    ("HORIZONTAL_LOGO_PNG", PIL.Image.open(drcutils.brand.HORIZONTAL_LOGO_PNG)),
    ("STACKED_LOGO_PNG", PIL.Image.open(drcutils.brand.STACKED_LOGO_PNG)),
    ("LOGO_ONLY_PNG", PIL.Image.open(drcutils.brand.LOGO_ONLY_PNG)),
    ("WHITE_PATTERN_PNG", PIL.Image.open(drcutils.brand.WHITE_PATTERN_PNG)),
    ("GREY_PATTERN_PNG", PIL.Image.open(drcutils.brand.GREY_PATTERN_PNG)),
    ("COLOR_PATTERN_PNG", PIL.Image.open(drcutils.brand.COLOR_PATTERN_PNG)),
]

# Iterate through subplots
subplot_index = 231
for im in brand_images:
    sp = matplotlib.pyplot.subplot(subplot_index)
    ax = sp.imshow(im[1])
    sp.set_title(im[0])
    matplotlib.pyplot.axis('off');
    subplot_index += 1

# Make sure it looks ok
fig.set_figwidth(10)
fig.set_dpi(200)
```
